```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from webdriver_manager.microsoft import EdgeChromiumDriverManager
from selenium.webdriver.edge.options import Options
import time
import json

def get_toto_prize():
    # Set up the Selenium WebDriver with headless Edge (runs without a GUI)
    options = Options()
    options.add_argument('--headless')  # Run in headless mode (no browser window)

    # Create the Edge driver
    driver = webdriver.Edge(service=Service(EdgeChromiumDriverManager().install()), options=options)
    
    url = "https://online.singaporepools.com/en/lottery"
    driver.get(url)
    print(driver)

    # Allow some time for the JavaScript to load content
    time.sleep(5)  # Adjust sleep time if necessary

    # Find the divs with data-component="Banner"
    banners = driver.find_elements_by_css_selector("div[data-component='Banner']")

    # Iterate through the banners to find the jackpot
    for banner in banners:
        data_prop_banner = banner.get_attribute("data-prop-banner")
        if data_prop_banner:
            # Parse the JSON from the data attribute
            banner_data = json.loads(data_prop_banner)

            # Check if the jackpot exists and its value
            jackpot = banner_data.get("jackpot", 0)
            if jackpot > 0:  # Adjust the condition as needed
                print(f"Found matching jackpot: SGD {jackpot:,}")
                driver.quit()  # Close the browser once done
                return jackpot

    driver.quit()  # Ensure the browser is closed if no jackpot is found
```

```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options
from webdriver_manager.microsoft import EdgeChromiumDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def get_toto_jackpot():
    options = Options()
    options.add_argument('--headless')  # Run in headless mode (remove if debugging)
    
    # Create the Edge driver
    driver = webdriver.Edge(service=Service(EdgeChromiumDriverManager().install()), options=options)

    url = "https://online.singaporepools.com/en/lottery"
    driver.get(url)

    # Wait for the main content to load
    try:
        jackpot_div = WebDriverWait(driver, 60).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "div.slab.slab--jackpot"))
        )
        jackpot_amount = jackpot_div.find_element(By.CSS_SELECTOR, "span.slab__text--highlight").text

        print(f"Jackpot Amount: {jackpot_amount}")
        return jackpot_amount
    except Exception as e:
        print(f"Error: {e}")

    driver.quit()

prize = get_toto_jackpot()
```
